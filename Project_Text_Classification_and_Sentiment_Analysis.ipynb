Load data from hugging_face

```
from datasets import load_dataset

myData = load_dataset("dair-ai/emotion")

myData
```

```
myData["train"]
```

```
text_label = myData["train"].features
text_label
```

```
text_label = myData["train"].features["label"].names
text_label
```

lets view myData with Panda

```
data_2_train = myData["train"]
data_pd = data_2_train.to_pandas()

data_pd
```

```

text_labels = []
text_label = myData["train"].features["label"].names

for label in data_pd["label"]:
  text_labels.append(text_label[label])

data_pd["text_labels"] = text_labels

data_pd
```



>Next step



```
import matplotlib.pyplot as plt

label_count = data_pd["text_labels"].value_counts(ascending=True)
label_count
```

```
import matplotlib.pyplot as plt

# plot some chart
plt.title("Emotion Graph")
plt.xlabel("Emotion")
plt.ylabel("Count")
plt.bar(label_count.index, label_count)
plt.show()
```
